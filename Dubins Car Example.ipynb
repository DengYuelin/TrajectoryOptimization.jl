```julia
using TrajectoryOptimization
using Plots
```

Load in Dubins car dynamics and predefined cost function parameters

```julia
model, obj = Dynamics.dubinscar;
```

Create constrained objective

```julia
u_min = [-1; -1]
u_max = [100; 100]
x_min = [0; -100; -100]
x_max = [1.0; 100; 100]
obj_con = TrajectoryOptimization.ConstrainedObjective(obj, u_min=u_min, u_max=u_max, x_min=x_min, x_max=x_max); # constrained objective
```

```julia
solver = Solver(model,obj_con,integration=:rk4,dt=0.1)

U = rand(solver.model.m,solver.N)

sol = TrajectoryOptimization.solve(solver,U)
println("Final state: $(sol.X[:,end])")
println("Final cost: $(sol_foh_con2.cost[sol_foh_con2.termination_index])")
```
